# Transforming Data

```python
#Fill unknown groups with unknown
df.funder.fillna('unknown', inplace=True)
df.installer.fillna('unknown', inplace=True)
df.subvillage.fillna('unknown', inplace=True)
df.scheme_management.fillna('unknown', inplace=True)
df.scheme_name.fillna('unknown', inplace=True)

testset.funder.fillna('unknown', inplace=True)
testset.installer.fillna('unknown', inplace=True)
testset.subvillage.fillna('unknown', inplace=True)
testset.scheme_management.fillna('unknown', inplace=True)
testset.scheme_name.fillna('unknown', inplace=True)
```

```python
#Fill T/F with False, likely false if not listed
df.permit.fillna('False', inplace=True)
df.public_meeting.fillna('False', inplace=True)

testset.permit.fillna('False', inplace=True)
testset.public_meeting.fillna('False', inplace=True)
```

```python
#Create bins for construction_year so the data can be grouped into time periods

bins = [1960., 1970., 1980., 1990., 2000., 2010.,2013.]
periods = ['1960-1969','1970-1979','1980-1989','1990-1999','2000-2009','2010-2013']

df['construction_periods'] = pd.cut(df['construction_year'], bins, labels=periods, right=False)
testset['construction_periods'] = pd.cut(testset['construction_year'], bins, labels=periods, right=False)
```

```python
df['construction_periods'] = df['construction_periods'].cat.add_categories('unknown')
df['construction_periods'].fillna('unknown', inplace =True) 

testset['construction_periods'] = testset['construction_periods'].cat.add_categories('unknown')
testset['construction_periods'].fillna('unknown', inplace =True) 
```

```python
df = df.drop('construction_year', axis=1)
testset = testset.drop('construction_year', axis=1)
```

Going to only include the values that are not unknown/0 anmd have >1000 value counts.

```python
def installer_cleaner(row):
  
  if row['installer'] == 'DWE':
    return 'DWE'
  elif row['installer'] == 'Government':
    return 'Government'
  elif row['installer'] == 'RWE':
    return 'RWE'
  elif row['installer'] == 'Commu':
    return 'Commu'
  elif row['installer'] == 'DANIDA':
    return 'Danida'
  else:
    return 'other'
```

```python
def funder_cleaner(row):
  
  if row['funder'] == 'Government of Tanzania':
    return 'Government of Tanzania'
  elif row['funder'] == 'Danida':
    return 'Danida'
  elif row['funder'] == 'Hesawa':
    return 'Hesawa'
  elif row['funder'] == 'Rwssp':
    return 'Rwssp'
  elif row['funder'] == 'World Bank':
    return 'World Bank'
  elif row['funder'] == 'Unicef':
    return 'Unicef'
  else:
    return 'other'
```

```python
df['installer'] = df.apply(lambda row: installer_cleaner(row), axis=1)
testset['installer'] = testset.apply(lambda row: installer_cleaner(row), axis=1)

df['funder'] = df.apply(lambda row: funder_cleaner(row), axis=1)
testset['funder'] = testset.apply(lambda row: funder_cleaner(row), axis=1)
```
